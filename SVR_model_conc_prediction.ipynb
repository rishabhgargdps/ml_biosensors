```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train_10 = pd.read_excel('10_train.xlsx', header = None)
df_train_10["Conc"] = 10
df_train_50 = pd.read_excel('50_train.xlsx', header = None)
df_train_50["Conc"] = 50
df_train_75 = pd.read_excel('75_train.xlsx', header = None)
df_train_75["Conc"] = 75
df_test = pd.read_excel('100_test.xlsx', header = None)
df_train_300 = pd.read_excel('300_train.xlsx', header = None)
df_train_300["Conc"] = 300
df_train_500 = pd.read_excel('500_train.xlsx', header = None)
df_train_500["Conc"] = 500
```

```python
dfList = [df_train_10, df_train_75, df_train_300, df_train_500]
df_train = pd.concat(dfList)
```

```python
# df_train = pd.read_excel("train.xlsx") # For only one instance of each conc
df_train
```

```python
df_test = pd.read_excel("50_test.xlsx", header = None)
df_test
```

```python
# SVR-RBF
X = df_train.iloc[:, df_train.columns != "Conc"].values
y = df_train["Conc"].values
from sklearn.svm import SVR
regressor = SVR(gamma= 0.0001, kernel= 'rbf', C= 50)
regressor.fit(X, y)
regressor.predict(df_test)
```

```python
# SVR-POLY
from sklearn.svm import SVR
regressor = SVR(gamma= 0.001, kernel= 'poly', C= 5)
regressor.fit(X, y)
regressor.predict(df_test)
```

```python
# XGBoost
# from xgboost import XGBRegressor
# regressor = XGBRegressor(objective="reg:squarederror")
# regressor.fit(X, y)
# regressor.predict(df_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth=25)
regressor.fit(X, y)
regressor.predict(df_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors = 4)
regressor.fit(X, y)
regressor.predict(df_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
regressor = AdaBoostRegressor(random_state=0, n_estimators=100)
regressor.fit(X, y)
regressor.predict(df_test)
```

```python
from sklearn.linear_model import BayesianRidge
regressor = BayesianRidge()
regressor.fit(X, y)
regressor.predict(df_test)
```
