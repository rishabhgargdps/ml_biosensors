```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./Li ion.csv')
```

```python
df
```

```python
X = df.iloc[:, df.columns != "label"].values
y = df["label"].values
```

```python
from sklearn.svm import SVR
X_train, X_test, y_train, y_test = train_test_split(X, y)
regressor = SVR(kernel= 'poly')
regressor.fit(X_train, y_train)
```

```python
res = pd.DataFrame(data = {'Prediction' : regressor.predict(X_test), 'Actual' : y_test})
```

```python
res["Difference"] = res["Actual"] - res["Prediction"]
```

```python
res.head(60)
```

```python
squared_error = res['Difference'].pow(2).sum()
squared_error
```
