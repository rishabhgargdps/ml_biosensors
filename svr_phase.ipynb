```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./ML_mat.csv')
```

```python
df
```

```python
X = df.iloc[:, df.columns != "KCl"].values
y = df["KCl"].values
```

```python
#Completely using the dataset for training
from sklearn.svm import SVR
regressor = SVR(gamma= 0.01, kernel= 'poly', C= 50)
regressor.fit(X, y)
print(regressor.predict([X[0]]), regressor.predict([X[1]]), regressor.predict([X[2]]), regressor.predict([X[3]]))
```

```python
#75% training 25% test data
from sklearn.svm import SVR
regressor = SVR(gamma= 0.01, kernel= 'poly', C= 50)
X_train = df.iloc[0:3, df.columns != "KCl"].values
y_train = df.iloc[0:3, df.columns == "KCl"].values.ravel()
regressor.fit(X_train, y_train)
print(regressor.predict([X[0]]), regressor.predict([X[1]]), regressor.predict([X[2]]), regressor.predict([X[3]]))
```

```python
#50% training 50% test data
from sklearn.svm import SVR
regressor = SVR(gamma= 0.01, kernel= 'poly', C= 50)
X_train = df.iloc[0:2, df.columns != "KCl"].values
y_train = df.iloc[0:2, df.columns == "KCl"].values.ravel()
regressor.fit(X_train, y_train)
print(regressor.predict([X[0]]), regressor.predict([X[1]]), regressor.predict([X[2]]), regressor.predict([X[3]]))
```

```python
#Preprocessing:- Min Max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = df.iloc[0:3, df.columns != "KCl"].values
y_train = df.iloc[0:3, df.columns == "KCl"].values.ravel()
X_train_scaled = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train_scaled)
regressor.fit(X_train, y_train)
X_test = df.iloc[3, df.columns != "KCl"].values
X_test_scaled = scaler.transform([X_test])
X_test = pd.DataFrame(X_test_scaled)
regressor.predict(X_test)
```
