```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./ML_mat_real_im.csv', header=None)
```

```python
df.rename(columns={122: "KCl"}, inplace = True)
```

```python
df
```

```python
X = df.iloc[:, df.columns != "KCl"].values
y = df["KCl"].values
```

```python
#Completely using the dataset for training
import xgboost
from xgboost import XGBRegressor
model = XGBRegressor()
model.fit(X, y)
print(model.predict([X[0]]), model.predict([X[1]]), model.predict([X[2]]), model.predict([X[3]]))
```

```python
#75% training 25% test data
model = XGBRegressor()
X_train = df.iloc[0:3, df.columns != "KCl"].values
y_train = df.iloc[0:3, df.columns == "KCl"].values.ravel()
model.fit(X_train, y_train)
print(model.predict([X[0]]), model.predict([X[1]]), model.predict([X[2]]), model.predict([X[3]]))
```

```python
#75% training 25% test data
model = XGBRegressor(n_estimators=500, max_depth=70)
X_train = df.iloc[0:3, df.columns != "KCl"].values
y_train = df.iloc[0:3, df.columns == "KCl"].values.ravel()
model.fit(X_train, y_train)
print(model.predict([X[0]]), model.predict([X[1]]), model.predict([X[2]]), model.predict([X[3]]))
```

```python
#50% training 50% test data
model = XGBRegressor(n_estimators=500, max_depth=10)
X_train = df.iloc[0:2, df.columns != "KCl"].values
y_train = df.iloc[0:2, df.columns == "KCl"].values.ravel()
model.fit(X_train, y_train)
print(model.predict([X[0]]), model.predict([X[1]]), model.predict([X[2]]), model.predict([X[3]]))
```

```python
#Preprocessing:- Min Max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = df.iloc[0:3, df.columns != "KCl"].values
y_train = df.iloc[0:3, df.columns == "KCl"].values.ravel()
X_train_scaled = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train_scaled)
model.fit(X_train, y_train)
X_test = df.iloc[3, df.columns != "KCl"].values
X_test_scaled = scaler.transform([X_test])
X_test = pd.DataFrame(X_test_scaled)
model.predict(X_test)
```
